# CTR simulation, Epsilon greedy algorithm

### setup

```python
#import packages
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

```python
#function for generating distributions
def gen_dist(list, n):
    return np.random.beta(list[0], list[1], size = (n,1))

def no_from_dist(list):
    return np.random.beta(list[0], list[1])
```

```python
#find the distribution with highest mean percentile
def Eps_decision(a):
    d = {float((i[0]/(i[0] + i[1]))) : i[2] for i in a}
    
    return d[max(d.keys())]
```

```python
ad_a_prior_par = ([80, 1000, 'a'])
ad_b_prior_par = ([90, 1000, 'b'])
ad_c_prior_par = ([120, 1000, 'c'])
Eps_decision([ad_a_prior_par, ad_b_prior_par, ad_c_prior_par])
```

```python
#randomly draw click or no click and update the distribution
def update_prior(b, epsilon):
    best_dist = Eps_decision(b)
    if (np.random.uniform(0,1,1) < epsilon):
        distributions = ([i[2] for i in b])
        distributions.remove(best_dist)
        rand_dist = random.choice(distributions)
        incr = (np.random.uniform(0,1,1) < eval('ad_' + rand_dist + '_CTR'))
        eval('ad_' + rand_dist + '_post_dist')[0] += incr
        eval('ad_' + rand_dist + '_post_dist')[1] +=  (1-incr)
    else:
        incr = (np.random.uniform(0,1,1) < eval('ad_' + best_dist + '_CTR'))
        eval('ad_' + best_dist + '_post_dist')[0] += incr
        eval('ad_' + best_dist + '_post_dist')[1] +=  (1-incr)
        
    return eval('ad_' + best_dist + '_post_dist')
    
```

### Define variables

```python
#true CTRs
ad_a_CTR = 0.061
ad_b_CTR = 0.020
ad_c_CTR = 0.060

#prior distributions
ad_a_prior_par = ([80, 1000, 'a'])
ad_b_prior_par = ([90, 1000, 'b'])
ad_c_prior_par = ([120, 1000, 'c'])

#posterior distributions that will be updated
ad_a_post_dist = ([80, 1000, 'a'])
ad_b_post_dist = ([90, 1000, 'b'])
ad_c_post_dist = ([120, 1000, 'c'])
```

### Simulation

```python
#update the prior distributions
impressions = 1000000
for j in range(impressions):
    update_prior([ad_a_post_dist, ad_b_post_dist, ad_c_post_dist], 0.1)
```
